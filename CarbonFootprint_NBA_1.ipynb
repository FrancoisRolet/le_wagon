# **Récupération de données depuis Baketball référence**

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Collect first page of match calendar
page = requests.get('https://www.basketball-reference.com/leagues/NBA_2023_games-october.html')
```

```python
# Create a BeautifulSoup object
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Pull all text from the BodyText div
data = soup.find(class_='left')
data
```

```python
# Pull all text from the BodyText div
data = soup.find(class_='right')
data
```

```python
soup.findAll('tr', limit=2)[0]
```

```python
soup.findAll('tr', limit=2)[1].findAll('th')
```

```python
# use getText()to extract the headers into a list
headers = [th.getText() for th in soup.findAll('tr', limit=2)[1].findAll('th')]
headers
```

```python
# get rows from table
rows = soup.findAll('tr')[0:]
rows_data = [[td.getText() for td in rows[i].findAll('td')]
                    for i in range(len(rows))]
rows_data
```
